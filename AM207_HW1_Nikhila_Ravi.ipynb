### Part A:
Your child has been randomly selected for Type I diabetes screening, using a highly accurate new test that boasts of a false positive rate of 1% and a false negative rate of 0%. The prevalence of of Type I diabetes in children is approximately 0.228%. Should your child test positive, what is the probability that he/she has Type I diabetes?

Should you be concerned enough to ask for further testing or treatment for your child?

Later, you read online that Type I diabetes is 6 times more prevalent in prematurely born children. If this statistic is true, what is the probability that your child, who is prematurely born, has Type I diabetes?

Given the new information, should you be concerned enough to ask for further testing or treatment for your child?

**Justify your decisions using your calculations.**

Let D be the event that the patient has diabetes
Let T be the event that the test is positive

$$ P(T | D^c) = 0.01 $$

$$ P(T^c|D^c) = 0.99 $$

$$ P(T^c|D) = 0 $$

$$ P(T|D) = 1 $$

$$ P(D) = 0.00288 $$

$$ P(D^c) = 1-P(D) = 1-0.00288 $$

$$ P(D|T) = \frac{P(T|D)P(D)}{P(T|D)P(D) + P(T|D^c)P(D^c)} $$

$$ P(D|T) = \frac{1*0.00288}{1*0.00288 + 0.01*(1-0.00288)} $$


```python
P = (1*0.00228)/(1*(0.00228) + 0.01*(1-0.00228))
print 'probabilty of disease: ', P
```

P(D) is now the probability of a child who is prematurely born having type I diabetes which is 6*P(D)

$$ P(D|T) = \frac{P(T|D)*6P(D)}{P(T|D)*6P(D) + P(T|D^c)P(D^c)} $$

```python
P = (0.99*6*0.00228)/(0.99*6*(0.00228) + 0.01*(1-0.00228))
print 'new probabilty of disease: ', P
```

Given this new higher probability, the child should be tested again for the disease.

## Problem 2: Maximum Likelihood Estimates

Suppose you observe the following data set $\mathbf{x}^{(0)} = (0.5, 2.5), \mathbf{x}^{(1)} = (3.2, 1.3), \mathbf{x}^{(2)} = (2.72, 5.84), \mathbf{x}^{(3)}= (10.047, 0.354)$. By convention, for any vector $\mathbf{x}$, we will denote the first component of $\mathbf{x}$ by $x_{1}$ and the second component by $x_{2}$. Suppose that the data is drawn from the same two-dimensional probability distribution with pdf $f_X$, that is, $\mathbf{x}^{(i)} \overset{iid}{\sim} f_X$, where
$$
f_X(\mathbf{x}) =  4\lambda_1^2 x_{1}x_{2} \mathrm{exp} \left\{-\lambda_0 (x^2_{1} + x^2_{2}) \right\}.
$$
You should assume that $\lambda_1, \lambda_0 > 0$ and that $f_X$ is supported on the nonnegative quandrant of $\mathbb{R}^2$ (i.e. $f_X$ is zero when either component is negative). 

### Part A:

As $f_X$ is a pdf it should satisfy the following condition:

$$
\int_{-\infty}^{+\infty} \int_{-\infty}^{+\infty}{f_X(\mathbf{x}) dx_1 dx_2} =  1
$$

where $x_1$ and $x_2$ are the two components of the $\mathbf{x}$ vector. 

As $f_X$ is only supported in the non-negative quandrant of $\mathbb{R}^2$ this simplifies to: 
$$
\int_{0}^{+\infty} \int_{0}^{+\infty} 4\lambda_1^2 x_{1}x_{2} \mathrm{exp} \left\{-\lambda_0 (x^2_{1} + x^2_{2}) \right\} dx_1 dx_2 = 1
$$

The double integral can be evaluated to find an expression for the unknown parameters $\lambda_0$ and $\lambda_1$:

$$
\int_{0}^{+\infty} \int_{0}^{+\infty} 4\lambda_1^2 x_{1}x_{2} \mathrm{exp} \left\{-\lambda_0 (x^2_{1} + x^2_{2}) \right\} dx_1 dx_2 = 1
$$

$$
\int_{0}^{+\infty} \frac{-2\lambda_1^{2}x_2e^{-\lambda_0(x_1^2)}}{\lambda_0}dx_1 =  1
$$

$$
\frac{\lambda_1^{2}}{\lambda_0^2} =  1
$$

$$
\therefore \quad \frac{\lambda_1}{\lambda_0} =  1
$$

This result can be used in the maximum likelihood estimation calculation. $\lambda_1$ can be subsituted by $\lambda_0$:

$$L(\lambda_0, \lambda_1)=∏_{i=1}^{n=3}(f_X(x_i∣\lambda_0))$$

$$log(L(\lambda_0, \lambda_1))=\sum_{i=1}^{n=3}log((4*\lambda_{0}^2x_{1}x_{2}\mathrm{exp} \left\{-\lambda_0 (x^{(i)^{2}}_{1} + x^{(i)^{2}}_{2}) \right\}))$$

$$log(L(\lambda_0, \lambda_1)=\sum_{i=1}^{n=3}log(4*\lambda_{0}^2x_{1}x_{2}) + \left\{-\lambda_0 (x^{(i)^{2}}_{1} + x^{(i)^{2}}_{2}) \right\}$$

$$\frac{\partial  log(L(\lambda_0, \lambda_1))}{\partial \lambda_0}=\sum_{i=1}^{n=3}[\frac{2}{\lambda_0} - (x^{(i)^{2}}_{1} + x^{(i)^{2}}_{2})]$$

By setting $\frac{\partial  log(L(\lambda_0, \lambda_1))}{\partial \lambda_0} = 0 $, solve for $\lambda_0$ :

$$ 0 = \frac{8}{\lambda_0} - [0.5^2 + 2.5^2 + 3.2^2 + 1.3^2 + 2.72^2 + 5.84^2 + 10.074^2 + 0.354^2]$$

$$ \lambda_0 = 0.04968$$

Therefore the pdf can be rewritten as:

$$
f_X(\mathbf{x}) =  4*0.002468 * x_{1}x_{2} \mathrm{exp} \left\{-0.04968 *(x^2_{1} + x^2_{2}) \right\}.
$$


```python
l = float(8)/(0.5**2 + 2.5**2 + 3.2**2 + 1.3**2 + 2.72**2 + 5.84**2 + 10.047**2 + 0.354**2)
print 'lambda_0', l
print 'lambda_0^2', l**2
```

## Visualising the pdf $f_X$

```python
import math
X = [0.5, 3.2, 2.72, 10.047]
Y = [2.5, 1.3, 5.84, 0.354]
xs = np.linspace(0, 12, 1000)
ys = np.linspace(0, 7, 1000)
xv, yv = np.meshgrid(xs, ys)
def pdf(x, y):
    return 4*(0.002468)*np.exp(-0.04968*(x**2 + y**2))*x*y
z = pdf (xv, yv)


fig, ax = plt.subplots(1, 1, figsize=(20, 10))
m = ax.contourf(xs, ys, z, 20, cmap=plt.cm.Greens)
ax.scatter(X, Y, s=80)
ax.scatter(xs, ys, c=z[1], s=0, cmap=m.cmap, vmin=m.vmin, vmax=m.vmax)
cb = fig.colorbar(m)
cb.set_label('$f(x_1, x_2)$', fontsize='xx-large')
ax.set_xlabel('$x_1$', fontsize='xx-large')
ax.set_ylabel('$y_2$', fontsize='xx-large')
ax.margins(0.05)
fig.tight_layout()
plt.show()
```
